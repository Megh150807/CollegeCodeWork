# FCFS, SJF, NON-PREEMPTIVE PRIORITY SCHEDULING

```python
class Process:
    def __init__(self, pid, bt, at, priority=0):
        self.pid = pid
        self.bt = bt
        self.at = at
        self.priority = priority
        self.ct = 0
        self.tat = 0
        self.wt = 0
        self.completed = False

def fcfs(processes):
    processes.sort(key=lambda x: x.at)
    
    current_time = 0
    total_wt = 0
    total_tat = 0
    
    print("\n--- FCFS Scheduling ---")
    print("PID\tAT\tBT\tCT\tTAT\tWT")
    
    for p in processes:
        if current_time < p.at:
            current_time = p.at
        
        p.ct = current_time + p.bt
        p.tat = p.ct - p.at
        p.wt = p.tat - p.bt
        
        current_time = p.ct
        total_wt += p.wt
        total_tat += p.tat
        
        print(f"P{p.pid}\t{p.at}\t{p.bt}\t{p.ct}\t{p.tat}\t{p.wt}")
        
    print(f"Avg WT: {total_wt/len(processes):.2f} | Avg TAT: {total_tat/len(processes):.2f}")

def sjf_np(processes):
    n = len(processes)
    completed = 0
    current_time = 0
    total_wt = 0
    total_tat = 0
    
    for p in processes: p.completed = False
        
    print("\n--- SJF (Non-Preemptive) ---")
    print("PID\tAT\tBT\tCT\tTAT\tWT")
    
    while completed != n:
        available = [p for p in processes if p.at <= current_time and not p.completed]
        
        if available:
            shortest = min(available, key=lambda x: x.bt)
            
            shortest.ct = current_time + shortest.bt
            shortest.tat = shortest.ct - shortest.at
            shortest.wt = shortest.tat - shortest.bt
            
            total_wt += shortest.wt
            total_tat += shortest.tat
            shortest.completed = True
            completed += 1
            current_time = shortest.ct
            
            print(f"P{shortest.pid}\t{shortest.at}\t{shortest.bt}\t{shortest.ct}\t{shortest.tat}\t{shortest.wt}")
        else:
            current_time += 1
            
    print(f"Avg WT: {total_wt/n:.2f} | Avg TAT: {total_tat/n:.2f}")

def priority_np(processes):
    n = len(processes)
    completed = 0
    current_time = 0
    total_wt = 0
    total_tat = 0
    
    for p in processes: p.completed = False
        
    print("\n--- Priority (Non-Preemptive) ---")
    print("PID\tAT\tBT\tPrio\tCT\tTAT\tWT")
    
    while completed != n:
        available = [p for p in processes if p.at <= current_time and not p.completed]
        
        if available:
            highest_prio = min(available, key=lambda x: x.priority)
            
            highest_prio.ct = current_time + highest_prio.bt
            highest_prio.tat = highest_prio.ct - highest_prio.at
            highest_prio.wt = highest_prio.tat - highest_prio.bt
            
            total_wt += highest_prio.wt
            total_tat += highest_prio.tat
            highest_prio.completed = True
            completed += 1
            current_time = highest_prio.ct
            
            print(f"P{highest_prio.pid}\t{highest_prio.at}\t{highest_prio.bt}\t{highest_prio.priority}\t{highest_prio.ct}\t{highest_prio.tat}\t{highest_prio.wt}")
        else:
            current_time += 1
            
    print(f"Avg WT: {total_wt/n:.2f} | Avg TAT: {total_tat/n:.2f}")

if __name__ == "__main__":
    n = int(input("Enter Number of Processes: "))
    procs = []
    print("Enter Process Details (BT AT Priority):")
    for i in range(n):
        bt, at, prio = map(int, input(f"P{i+1}: ").split())
        procs.append(Process(i+1, bt, at, prio))

    print("\n1. FCFS\n2. SJF (Non-Preemptive)\n3. Priority (Non-Preemptive)")
    choice = int(input("Enter Choice: "))

    if choice == 1: fcfs(procs)
    elif choice == 2: sjf_np(procs)
    elif choice == 3: priority_np(procs)
```

# PREEMPTIVE SCHEDULING (SRTF, PRIORITY, ROUND ROBIN) 

```python
import sys

class Process:
    def __init__(self, pid, bt, at, priority=0):
        self.pid = pid
        self.bt = bt
        self.at = at
        self.priority = priority
        self.rem_bt = bt
        self.ct = 0
        self.tat = 0
        self.wt = 0

def srtf(processes):
    n = len(processes)
    current_time = 0
    completed = 0
    total_wt = 0
    total_tat = 0

    for p in processes: p.rem_bt = p.bt
        
    print("\n--- SRTF (Preemptive SJF) ---")
    print("PID\tAT\tBT\tCT\tTAT\tWT")
    
    while completed != n:
        available = [p for p in processes if p.at <= current_time and p.rem_bt > 0]
        
        if not available:
            current_time += 1
            continue
            
        shortest = min(available, key=lambda x: x.rem_bt)
        
        shortest.rem_bt -= 1
        current_time += 1
        
        if shortest.rem_bt == 0:
            completed += 1
            shortest.ct = current_time
            shortest.tat = shortest.ct - shortest.at
            shortest.wt = shortest.tat - shortest.bt
            total_wt += shortest.wt
            total_tat += shortest.tat
            
            print(f"P{shortest.pid}\t{shortest.at}\t{shortest.bt}\t{shortest.ct}\t{shortest.tat}\t{shortest.wt}")
            
    print(f"Avg WT: {total_wt/n:.2f} | Avg TAT: {total_tat/n:.2f}")

def priority_preemptive(processes):
    n = len(processes)
    current_time = 0
    completed = 0
    total_wt = 0
    total_tat = 0
    
    for p in processes: p.rem_bt = p.bt
        
    print("\n--- Preemptive Priority ---")
    print("PID\tAT\tBT\tPrio\tCT\tTAT\tWT")
    
    while completed != n:
        available = [p for p in processes if p.at <= current_time and p.rem_bt > 0]
        
        if not available:
            current_time += 1
            continue
            
        highest_prio = min(available, key=lambda x: x.priority)
        
        highest_prio.rem_bt -= 1
        current_time += 1
        
        if highest_prio.rem_bt == 0:
            completed += 1
            highest_prio.ct = current_time
            highest_prio.tat = highest_prio.ct - highest_prio.at
            highest_prio.wt = highest_prio.tat - highest_prio.bt
            total_wt += highest_prio.wt
            total_tat += highest_prio.tat
            
            print(f"P{highest_prio.pid}\t{highest_prio.at}\t{highest_prio.bt}\t{highest_prio.priority}\t{highest_prio.ct}\t{highest_prio.tat}\t{highest_prio.wt}")
            
    print(f"Avg WT: {total_wt/n:.2f} | Avg TAT: {total_tat/n:.2f}")

def round_robin(processes, tq):
    n = len(processes)
    current_time = 0
    completed = 0
    total_wt = 0
    total_tat = 0
    
    for p in processes: p.rem_bt = p.bt
    processes.sort(key=lambda x: x.at)
    
    queue = []
    queue = [p for p in processes if p.at <= current_time]
    
    print(f"\n--- Round Robin (TQ={tq}) ---")
    print("PID\tAT\tBT\tCT\tTAT\tWT")
    
    in_queue = set([p.pid for p in queue])
    
    while completed != n:
        if not queue:
            current_time += 1
            new_arrivals = [p for p in processes if p.at <= current_time and p.rem_bt > 0 and p.pid not in in_queue]
            for p in new_arrivals:
                queue.append(p)
                in_queue.add(p.pid)
            continue
            
        current_p = queue.pop(0)
        in_queue.remove(current_p.pid)
        
        time_slice = min(tq, current_p.rem_bt)
        
        current_p.rem_bt -= time_slice
        current_time += time_slice
        
        new_arrivals = [p for p in processes if p.at <= current_time and p.rem_bt > 0 and p.pid != current_p.pid and p.pid not in in_queue and p not in queue]
        for p in new_arrivals:
            queue.append(p)
            in_queue.add(p.pid)
            
        if current_p.rem_bt > 0:
            queue.append(current_p)
            in_queue.add(current_p.pid)
        else:
            completed += 1
            current_p.ct = current_time
            current_p.tat = current_p.ct - current_p.at
            current_p.wt = current_p.tat - current_p.bt
            total_wt += current_p.wt
            total_tat += current_p.tat
            print(f"P{current_p.pid}\t{current_p.at}\t{current_p.bt}\t{current_p.ct}\t{current_p.tat}\t{current_p.wt}")

    print(f"Avg WT: {total_wt/n:.2f} | Avg TAT: {total_tat/n:.2f}")

if __name__ == "__main__":
    n = int(input("Enter Number of Processes: "))
    procs = []
    print("Enter Process Details (BT AT Priority):")
    for i in range(n):
        bt, at, prio = map(int, input(f"P{i+1}: ").split())
        procs.append(Process(i+1, bt, at, prio))

    print("\n1. Round Robin\n2. SRTF\n3. Preemptive Priority")
    choice = int(input("Enter Choice: "))

    if choice == 1: 
        tq = int(input("Enter Time Quantum: "))
        round_robin(procs, tq)
    elif choice == 2: srtf(procs)
    elif choice == 3: priority_preemptive(procs)
```

# BANKER'S ALGORITHM

```python
def bankers_algorithm():
    n = int(input("Enter number of processes: "))
    m = int(input("Enter number of resources: "))
    
    alloc = []
    print("Enter Allocation Matrix (row by row):")
    for _ in range(n):
        alloc.append(list(map(int, input().split())))
        
    max_res = []
    print("Enter Max Matrix (row by row):")
    for _ in range(n):
        max_res.append(list(map(int, input().split())))
        
    avail = list(map(int, input("Enter Available Resources: ").split()))
    
    need = [[0]*m for _ in range(n)]
    for i in range(n):
        for j in range(m):
            need[i][j] = max_res[i][j] - alloc[i][j]
            
    finish = [False] * n
    safe_seq = []
    work = avail[:]
    
    count = 0
    while count < n:
        found = False
        for i in range(n):
            if not finish[i]:
                if all(need[i][j] <= work[j] for j in range(m)):
                    for k in range(m):
                        work[k] += alloc[i][k]
                    
                    safe_seq.append(i)
                    finish[i] = True
                    found = True
                    count += 1
                    
        if not found:
            print("\nSystem is NOT in a safe state (Deadlock detected).")
            print("Deadlocked processes:", end=" ")
            for i in range(n):
                if not finish[i]: print(f"P{i}", end=" ")
            print()
            return

    print("\nSystem is in a SAFE State.")
    print("Safe Sequence: ", end="")
    print(" -> ".join([f"P{p}" for p in safe_seq]))

if __name__ == "__main__":
    bankers_algorithm()
```
